```python
#Bar chart
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
used_cars=pd.read_csv('bar_chart_data.csv')
plt.bar(x=used_cars['Brand'],height=used_cars['Cars Listings'],color='red')
plt.xticks(rotation=45,fontsize=13)
plt.yticks(fontsize=13)
plt.title('cars listings by brand',fontsize=16,fontweight='bold')
plt.ylabel('number of listings',fontsize=13)
plt.show()
plt.savefig('used cars plot.png')
```

```python
#bar chart
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('colorblind')
cities=pd.read_csv('bar_chart_homework_data.csv')
plt.figure(figsize=(10,8))
plt.bar(x=cities['Cities'],height=cities['Frequency'],color='Gold')
plt.xticks(fontsize=13,fontweight='bold')
plt.yticks(fontsize=13)
plt.title('cities propotional to resturants')
plt.xlabel('Cities',fontweight='bold')
plt.ylabel('Frequency',fontweight='bold')
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set_palette('colorblind')
df_engine=pd.read_csv('pie_chart_data.csv')
plt.figure(figsize=(10,8))
plt.pie(df_engine['Number of Cars'],
       labels=df_engine['Engine Fuel Type'].values,
       autopct='%.2f%%',
       textprops={'size':'x-large','fontweight':'bold','rotation':45,'color':'w'})
plt.legend()
plt.title('cars by engine type',fontsize=18,fontweight='bold')
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set_palette('colorblind')
engine=pd.read_csv('pie_chart_data.csv')
plt.figure(figsize=(10,8))
plt.pie(engine['Number of Cars'],
       labels=df_engine['Engine Fuel Type'],
        autopct='%.2f%%',
       textprops={'size':'x-large','fontweight':'bold','rotation':45,'color':'black'})
plt.legend()
plt.title('cars by engine type',fontsize=18,fontweight='bold')
plt.show()
```

```python
#stacked area chart
engine_type=pd.read_csv('stacked_area_chart_data.csv')
color=['y','r','b']
label=['Diesel','Petrol','Gas']
sns.set_style('white')
plt.figure(figsize=(12,6))
plt.stackplot(engine_type['Year'],
              engine_type['Diesel'],
              engine_type['Petrol'],
              engine_type['Gas'],
             colors=color,
             edgecolor='none')
plt.xticks(engine_type['Year'],rotation=45)
plt.legend(labels=label,loc='upper right')
plt.ylabel('No of cars')
plt.title('Popularity of fuel type from 1982 to 2016',fontsize='x-large',fontweight='bold')
sns.despine(right=True,top=True)
plt.show()
```

```python
#stacked area chart
flights=pd.read_csv('area_chart_homework_data.csv')
color=['red','blue','green']
label=['International','Domestic']
sns.set_style('darkgrid')
plt.figure(figsize=(18,8))
plt.stackplot(flights['Date'],
             flights['International'],
             flights['Domestic'],
             flights['Services'],
             colors=color,
             edgecolor='none')
plt.xticks(flights['Date'],rotation=45)
plt.legend(labels=label,loc='upper right')
plt.xlabel('Date')
plt.title('Graph representing flights route and service',fontsize='x-large',fontweight='bold')
sns.despine(right=True,top=True)
plt.show()
```

```python
#Line chart 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
GSPC=pd.read_csv('line chart returns1020.csv')
GSPC['new date']=pd.to_datetime(GSPC['Date'])
labels=['gspc','ftse']
plt.figure(figsize=(20,8))
plt.plot(GSPC['new date'],GSPC['GSPCRet'],color='r')
plt.plot(GSPC['new date'],GSPC['FTSERet'],color='b')
plt.title('s & p vs ftse returns from 2008 to 2010',fontsize=14,fontweight='bold')
plt.legend(labels=labels)
plt.ylabel('returns',fontsize=18,fontweight='bold')
plt.xlabel('date',fontsize=18,fontweight='bold')
plt.show()
```

```python
#line chart specified time frame
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
sp_ftse=pd.read_csv('line_chart_data.csv')
sp_ftse['new_date']=pd.to_datetime(sp_ftse["Date"])
mask = (sp_ftse['new_date'] >= '2008-07-01') & (sp_ftse['new_date'] <= '2008-12-31')
sp_ftse2008 = sp_ftse.loc[mask]
labels=['s&p 500','ftse 100']
plt.figure(figsize=(20,8))
plt.plot(sp_ftse2008['new_date'],sp_ftse2008['GSPC500'],color='r')
plt.plot(sp_ftse2008['new_date'],sp_ftse2008['FTSE100'],color='b')
plt.title('s & p vs ftse returns from 2008 to 2010',fontsize=14,fontweight='bold')
plt.ylabel('returns',fontsize=18,fontweight='bold')
plt.xlabel('date',fontsize=18,fontweight='bold')
plt.legend(labels=labels,fontsize='x-large')
plt.show()
```

```python
#histogram
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style('white')
df_real_estate=pd.read_csv('histogram_data.csv')
plt.figure(figsize=(8,6))
plt.hist(df_real_estate['Price'],
        bins=8,
        color='b')
plt.title('Distribution of real estate prices',fontsize=14,fontweight='bold')
plt.xlabel('price in (000.$)')
plt.ylabel('No of properties')
sns.despine(top=True,right=True)
plt.show()
```
